```python
#importing the library
from google.colab.patches import cv2_imshow
import cv2
```

# Adding the image

```python
#loading the respective image
my_image = cv2.imread("/content/drive/My Drive/LGMVIP-DataScience-main/LGMVIP-Task-1/lexus.jpg")

scale = 20 
width = int(my_image.shape[1] * scale / 100)
height = int(my_image.shape[0] * scale / 100)
dimension = (width, height)
  
# resize image
resized = cv2.resize(my_image, dimension, interpolation = cv2.INTER_AREA)
cv2_imshow(resized)
```

```python
#conversion to gray scale image
image_in_gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_in_gray)
```

```python
#inverting the grayscale image or transforming into negative using bitwise_not operator
image_invert = 255-image_in_gray
cv2_imshow(image_invert)
```

```python
#blurring the image using gaussian function
blurred_image = cv2.GaussianBlur(image_invert, (21, 21), 0)
cv2_imshow(blurred_image)
```

```python
#inverting the blurred image
blurred_invert = 255-blurred_image
cv2_imshow(blurred_invert)
```

# Final Pencil Sketch

```python
#final sketch
final_sketch = cv2.divide(image_in_gray, blurred_invert, scale=240.0)
cv2_imshow(final_sketch)
```
